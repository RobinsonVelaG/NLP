#### Extracción de los tuits de Ernesto Samper

```python
def get_all_tweets(screen_name):
#Este método solo tiene permitido descargar máximo los ultimos 3240 tweets del usuario

    auth = tweepy.OAuthHandler(keys['api_key'], keys['api_secret_key'])
    auth.set_access_token(keys['access_token'], keys['access_token_secret'])
    api = tweepy.API(auth, wait_on_rate_limit=True)
    
    alltweets = []  

    new_tweets = api.user_timeline(screen_name = screen_name,count=200)
    
    #esto permite traer los tweets más recientes
    alltweets.extend(new_tweets)
    
    #guardar el id del tweet más viejo menos uno
    oldest = alltweets[-1].id - 1
    
    #este comando es para extraer tweets hasta que no queden más para extraer
    while len(new_tweets) > 0:
        print(f"getting tweets before {oldest}")
        
        #todas las solicitudes subsecuentes usan el parametro max_id para prevenir duplicados
        new_tweets = api.user_timeline(screen_name = screen_name,count=200,max_id=oldest)
        
        #guardo los tweets más recientes
        alltweets.extend(new_tweets)
        
        #actualiza el id del tweet más viejo menos uno
        oldest = alltweets[-1].id - 1
        
        print(f"...{len(alltweets)} tweets downloaded so far")
    
    #transforma los tweets de tweepy en una matriz de dos dimensiones que va a llenar el csv 
    outtweets = [[tweet.id_str, tweet.created_at, tweet.text] for tweet in alltweets]
    
    #escribe el csv  
    with open(f'new_{screen_name}_tweets.csv', 'w', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(["id","created_at","text"])
        writer.writerow(outtweets)
    
    pass


if __name__ == '__main__':
    
#escribir entre comillas el nombre del usuario twitter para descargar los tuits
    get_all_tweets("ernestosamperp")

    
    
## el codigo anterior fue tomado y adapatado para el proyecto de  NLP de https://gist.github.com/yanofsky/5436496 
## A script to download all of a user's tweets into a csv
## usuario github: David Yanofsky
## Se trata de un software libre y gratuito de dominio público.
```

```python
def pre_procesado(texto):
    texto = re.sub(r"[\W\d_]+", " ", texto.lower())
    texto = [palabra for palabra in texto.split() if palabra not in stopwords_sp]
    texto = " ".join(texto)
    return texto

samper['pre-procesado'] = samper['text'].apply(lambda texto: pre_procesado(texto))
samper
```

```python
def get_stemmed_text(corpus):
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in corpus]

samper['stemmedtext'] = get_stemmed_text(samper['pre-procesado'])
samper
```

# Frecuencia de palabras

```python
def limpiar_tokenizar(texto):

    # Se convierte todo el texto a minúsculas
    nuevo_texto = texto.lower()
    # Eliminación de páginas web (palabras que empiezan por "http")
    nuevo_texto = re.sub('http\S+', ' ', nuevo_texto)
    # Eliminación de signos de puntuación
    regex = '[\\!\\"\\#\\$\\%\\&\\\'\\(\\)\\*\\+\\,\\-\\.\\/\\:\\;\\<\\=\\>\\?\\@\\[\\\\\\]\\^_\\`\\{\\|\\}\\~]'
    nuevo_texto = re.sub(regex , ' ', nuevo_texto)
    # Eliminación de números
    nuevo_texto = re.sub("\d+", ' ', nuevo_texto)
    # Eliminación de espacios en blanco múltiples
    nuevo_texto = re.sub("\\s+", ' ', nuevo_texto)
    # Tokenización por palabras individuales
    nuevo_texto = nuevo_texto.split(sep = ' ')
    # Eliminación de tokens con una longitud < 2
    nuevo_texto = [token for token in nuevo_texto if len(token) > 1]
    
    return(nuevo_texto)
```

```python
# Se aplica la función de limpieza y tokenización a cada tweet
# ==============================================================================
tweets['texto_tokenizado'] = tweets['texto'].apply(lambda x: limpiar_tokenizar(x))
tweets[['texto', 'texto_tokenizado']].head()
```

```python
# Unnest de la columna texto_tokenizado
# ==============================================================================
tweets_tidy = tweets.explode(column='texto_tokenizado')
tweets_tidy = tweets_tidy.drop(columns='texto')
tweets_tidy = tweets_tidy.rename(columns={'texto_tokenizado':'token'})
tweets_tidy.head(3)
```

# Correlación entre autores

```python
tweets_pivot = tweets_tidy.groupby(["Usuario","token"])["token"] \
                .agg(["count"]).reset_index() \
                .pivot(index = "token" , columns="Usuario", values= "count")
tweets_pivot.columns.name = None
```

```python
# Test de correlación (coseno) por el uso y frecuencia de palabras
from scipy.spatial.distance import cosine

def similitud_coseno(a,b):
    distancia = cosine(a,b)
    return 1-distancia

tweets_pivot.corr(method=similitud_coseno)
```

```python
# Gráfico de correlación
# las dos personas con mayor similitud (Samper y Santos 0.6586)
f, ax = plt.subplots(figsize=(6, 4))
temp = tweets_pivot.dropna()
sns.regplot(
    x  = np.log(temp.ernestosamperp),
    y  = np.log(temp.JuanManSantos),
    scatter_kws =  {'alpha': 0.05},
    ax = ax
);
for i in np.random.choice(range(temp.shape[0]), 100):
    ax.annotate(
        text  = temp.index[i],
        xy    = (np.log(temp.ernestosamperp[i]), np.log(temp.JuanManSantos[i])),
        alpha = 0.7
    )
```
