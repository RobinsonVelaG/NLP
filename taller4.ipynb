```python
import re
import pandas as pd
from collections import Counter

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from sklearn.feature_extraction.text import CountVectorizer
```

# Punto 1: Pre-Procesamiento

- `[5 pts]` Leer el archivo `dialogos.csv` usando `pandas`
- `[10 pts]` Crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números y sin palabras vacias
- `[7 pts]` Usando la nueva columna, ¿cuáles son las 10 palabras más usadas? 

```python
# Leer el archivo dialogos.csv usando pandas
dialogos = pd.read_csv('D:\KonradLorenz\D4298 G51 Electiva NLP 2021-I\Taller4\dialogos.csv', sep=';')
dialogos
```

```python
# Crear una nueva columna con el texto en minúscula
dialogos['pre-procesado'] = dialogos['Locución'].str.lower()
#dialogos['pre-procesado'] = dialogos['pre-procesado'].replace(regex=r'[\W\d]', value=' ')
#dialogos['pre-procesado'] = dialogos['pre-procesado'].str.strip()
#dialogos['pre-procesado'] = dialogos['pre-procesado'].str.split()
#dialogos['pre-procesado'] = [palabra for palabra in dialogos['pre-procesado'] if palabra not in stopwords_sp]
#dialogos
```

```python
# Sin caracteres especiales ni números y sin palabras vacias
cont=0
for filas in dialogos['Locución']:
    ns = filas.lower()
    ns = re.sub(r'[\W\d_]+', ' ', ns)
    token = ns.split()
    cadena = ''
    for filas2 in token:
        if filas2 in stopwords_sp:
            cadena += ''
            continue  
        else : 
            cadena += ' '+filas2 
    dialogos['pre-procesado'][cont] = cadena
    cadena = ''
    cont += 1
    if(cont==959):
        break
    continue
    
dialogos
```

```python
# Usando la nueva columna, ¿cuáles son las 10 palabras más usadas?
from collections import Counter
Counter(' '.join(dialogos['pre-procesado']).split()).most_common(10)
```
